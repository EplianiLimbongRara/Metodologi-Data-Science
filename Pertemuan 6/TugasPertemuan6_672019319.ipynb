JAWABAN NOMOR 3 Pada kasus ini saya memilih menggunakan Bar Chart, karena Bar Chart cocok digunakan untuk membandingkan nilai antar beberapa kategori untuk menunjukkan nilai terkecil dan terbesar dari suatu data, sehingga dari kasus ini diperoleh bahwa :
1. Popularitas Relatif Makanan Penutup yaitu Ice Cream
2. Berdasarkan presentasi dari grafik tersebut, maka dapat ditarik kesimpulan bahwa 3 makanan penutup yang harus disingkirkan adalah Kue Wajik, Pastel, dan Puding Vanila 
Keterangan : Warna Merah (Menu penutup dengan popularitas relatif) dan Warna Kuning (Menu penutup yang paling kurang diminati, sehingga dapat disingkirkan)  

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

datapenjualan_makananpenutup =['Donat',
                                'Pastel',
                                'Kue Coklat',
                                'Ice Cream',
                                'Puding Vanila',
                                'Brownies',
                                'Puding Strawberi',
                                'Puding Coklat',
                                'Ice Cream Nutela',
                                'Kue Coklat-Keju',
                                'Kue Wajik',
                                'Kue Sus','Mochi']
penjualan=(14, 5, 12, 19, 6, 8, 12, 9, 10, 17, 2, 9, 13)

df = pd.DataFrame({'Menu':datapenjualan_makananpenutup, 'Sale':penjualan})
df.sort_values(by='Sale',inplace = True, ascending = False)

kor_x = np.arange(len(df)) 
warna = ['#0000FF' for _ in range(len(df))]
warna [0] = '#FF0000'
warna [-1] = '#FFFF00'
warna [-2] = '#FFF000'
warna [-3] = '#FFF000'

plt.figure(figsize=(18,10))
plt.bar(kor_x, df['Sale'], tick_label=df['Menu'], color=warna)
plt.title('Data Penjualan Makanan Penutup Kafe Biru \n')
plt.xticks(rotation=90)
plt.xlabel('Nama Makanan Penutup')
plt.ylabel('Jumlah Penjualan')
plt.show()
```
